```python
import numpy as np
import scipy as sp
from scipy import linalg, integrate, optimize, stats, interpolate

matrix = np.array([[4, 7], [2, 6]])
inverse_matrix = linalg.inv(matrix)
print("Inverse of matrix:")
print(inverse_matrix)

def func(x):
    return x**2

result, error = integrate.quad(func, 0, 1)
print("\nNumerical Integration result:")
print(result)

def objective_function(x):
    return x**2 + 5*x + 6

opt_result = optimize.minimize(objective_function, 0)
print("\nOptimization result:")
print(opt_result)

data = np.random.randn(1000)
mean = stats.tmean(data)
std_dev = stats.tstd(data)
print("\nStatistical Functions:")
print(f"Mean: {mean}, Standard Deviation: {std_dev}")

x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])
f = interpolate.interp1d(x, y, kind='linear')
print("\nInterpolated value at x = 2.5:")
print(f(2.5))
```

```python
import tkinter as tk

def submit_action():
    name = name_entry.get()
    email = email_entry.get()
    print(f"Name: {name}")
    print(f"Email: {email}")

def reset_action():
    name_entry.delete(0, tk.END)
    email_entry.delete(0, tk.END)

window = tk.Tk()
window.title("Window Wizard")

name_label = tk.Label(window, text="Name:")
name_label.grid(row=0, column=0, padx=10, pady=10)

email_label = tk.Label(window, text="Email:")
email_label.grid(row=1, column=0, padx=10, pady=10)

name_entry = tk.Entry(window)
name_entry.grid(row=0, column=1, padx=10, pady=10)

email_entry = tk.Entry(window)
email_entry.grid(row=1, column=1, padx=10, pady=10)

submit_button = tk.Button(window, text="Submit", command=submit_action)
submit_button.grid(row=2, column=0, padx=10, pady=10)

reset_button = tk.Button(window, text="Reset", command=reset_action)
reset_button.grid(row=2, column=1, padx=10, pady=10)

window.mainloop()
```
